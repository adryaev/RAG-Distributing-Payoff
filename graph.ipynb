```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Read your corrected CSV file into a DataFrame
# You can change the filename to 'dcm_counts_corrected.csv' to visualize the other dataset
dcs_counts = 'dcs_counts.csv'
dcs_file = pd.read_csv(dcs_counts)
dcm_counts = 'dcm_counts.csv'
dcm_file = pd.read_csv(dcm_counts)

display(dcs_file.head())
top_n = 6
top_domains = dcs_file.sort_values(by='Count', ascending=False).head(top_n)
# --- Using the same 'top_domains' DataFrame from the previous step ---

# Create the pie chart
plt.figure(figsize=(5, 5))
plt.pie(top_domains['Count'], labels=top_domains['Website'], autopct='%1.1f%%', startangle=140)

# Add a title
plt.title(f'Top {top_n} Cited Websites couting Single Citation', fontsize=16)

# Equal aspect ratio ensures that pie is drawn as a circle.
plt.axis('equal') 

# Display the plot
plt.show()

display(dcm_file.head())
top_n = 6
top_domains = dcm_file.sort_values(by='Count', ascending=False).head(top_n)
# --- Using the same 'top_domains' DataFrame from the previous step ---
# Create the pie chart
plt.figure(figsize=(5, 5))
plt.pie(top_domains['Count'], labels=top_domains['Website'], autopct='%1.1f%%', startangle=140)

# Add a title
plt.title(f'Top {top_n} Cited Websites couting Multiple Citation', fontsize=16)

# Equal aspect ratio ensures that pie is drawn as a circle.
plt.axis('equal') 

# Display the plot
plt.show()




top_n = 25
# Sort the DataFrame by 'Count' and select the top N
top_domains = dcs_file.sort_values(by='Count', ascending=False).head(top_n)

# --- 3. Plotting ---

# Set a visually appealing style and a good figure size
sns.set_theme(style="whitegrid")
plt.figure(figsize=(12, 10)) # Made it a bit taller for 20 items

# Create the horizontal bar plot using a vibrant 'viridis' color palette
barplot = sns.barplot(x='Count', y='Website', data=top_domains, palette='viridis')

# Add data labels (the exact count) to the end of each bar for precision
for index, value in enumerate(top_domains['Count']):
    plt.text(value, index, f' {value}', va='center')

# Add a title and clear labels
plt.title(f'Top {top_n} Most Cited Websites', fontsize=18, weigst='bold')
plt.xlabel('Number of Citations', fontsize=12)
plt.ylabel('Website Domain', fontsize=12)

# Adjust plot limits to make space for the data labels
plt.xlim(0, top_domains['Count'].max() * 1.1) 

# Ensure a clean layout
plt.tight_layout()

# Display the final plot
plt.show()

```
